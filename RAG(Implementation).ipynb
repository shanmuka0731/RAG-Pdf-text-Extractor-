```python
def ask_llm(prompt):
    response = ollama.chat(
        model=model,
        messages=[
            {
                "role": "user",
                "content": prompt,
            }
        ],
        stream=True
    )

    for res in response:
        print(res["message"]["content"], end="", flush=True)
```

```python
while True:
    retriever=vector_store.as_retriever()
    query = input("Enter the query:")
    context = retriever.invoke(query)
    prompt=f"""You are an intelligent RAG assistant.answer the questions asked by the user with precise context and make it clear and understandable. And if you found any sort of topic that is not understandble just respond politely that "Sorry I couldn"t help you with this." 
           question:{query}
           context:{context}
           """
    ask_llm(prompt)
```

```python
def clear_vector_store(vector_store):
    """Deletes all documents from the current collection."""
    all_docs = vector_store.get()
    ids = all_docs.get("ids", [])
    
    if ids:
        vector_store.delete(ids=ids)
        print(f"--- Successfully deleted {len(ids)} documents from the vector store. ---")
    else:
        print("--- Vector store is already empty. ---")
```

```python
clear_vector_store(vector_store)
```
