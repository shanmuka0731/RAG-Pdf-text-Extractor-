```python
import ollama
from langchain_docling.loader import DoclingLoader
model="llama3:8b"
FILE_PATH=r"C:\Users\Shanmuka\Desktop\theMOVIE.pdf"
loader=DoclingLoader(file_path=FILE_PATH)
docs=loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=60)
text=text_splitter.split_documents(docs)
print(text[0])
print(text[1])
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="BAAI/bge-large-en-v1.5")
```

```python
pip install langchain-chroma
```

```python
from langchain_chroma import Chroma
vector_store=Chroma(
    collection_name="sample_RAG",
    embedding_function=embeddings,
    persist_directory="./chroma-langchain_db",
)
print(vector_store._collection.count())
```

```python
pip install langchain-community
```

```python
from langchain_community.vectorstores.utils import filter_complex_metadata
text=filter_complex_metadata(text)
vector_store.add_documents(text)
```

```python
vector_store._collection.count()
```

```python
vector_store._collection.get(ids=['2863586c-aebe-4ea0-a138-37c5adee9295'],include=["embeddings","documents"])
```

```python
retriever=vector_store.as_retriever()
```

```python
query="Where the story Begins?"
retriever_docs=retriever.invoke(query)
context="/n/n".join([doc.page_content for doc in retriever_docs])
print(context)
```
